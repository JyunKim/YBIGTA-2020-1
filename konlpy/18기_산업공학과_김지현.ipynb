```python
from konlpy.corpus import kolaw
%matplotlib inline 
```

```python
korealaw = kolaw.open('constitution.txt').read()
```

```python
klawwc = WordCloud(font_path='./data/NanumBarunGothic.ttf',
                   background_color = "white",
                   collocations = False,
                   max_font_size = 80,
                   width = 600,
                   height = 400)

klawwc.generate(korealaw)
```

```python
klawwc.to_image()
```

```python
klawwc.to_file('18기_산업공학과_김지현.png')
```
